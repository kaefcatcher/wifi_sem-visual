# WiFi своими руками или как мы сеть строили

## Первая практическая задачка:
Есть несколько островов на расстоянии 300 км друг от друга. Тянуть кабели дорого, поэтому нужно строить систему беспроводной связи. **Основная задача**: Обеспечить гарантированную доставку пакетов со нескольких терминалов на центральный компьютер. Делим ли очередь передачи на несколько очередей, если:
* Пользовательские данные поступают на центральный компьютер, образуя пуассоновский поток:
$$ P(k; \lambda) = \frac{\lambda^k e^{-\lambda}}{k!} $$
    * P(k;λ) — вероятность того, что случится k событий
    * λ — среднее количество событий за фиксированный интервал времени;
    * k — количество событий;
    * e — основание натурального логарифма (примерно равно 2.71828);

* Отброшенные из-за ошибок передачи пакеты передаются повторно, также образуя пуассоновский поток

* Все пакеты имеют одинаковую длину и передаются одинаковое время

* В сети находится бесконечное количество удаленных терминалов (то есть если некий терминал уже передает данные, это никак не влияет на вероятность передачи данных другими терминалами)

### Доказательство о неразделении трафика в сети ALOHA на несколько очередей

Вообще говоря, эта задача относится к теории систем массового обслуживания. (http://repo.ssau.ru/bitstream/Uchebnye-izdaniya/Sistemy-i-seti-massovogo-obsluzhivaniya-91842/1/%D0%A0%D0%BE%D0%BC%D0%B0%D0%BD%D0%B5%D0%BD%D0%BA%D0%BE%20%D0%92.%D0%90.%20%D0%A1%D0%B8%D1%81%D1%82%D0%B5%D0%BC%D1%8B%20%D0%B8%20%D1%81%D0%B5%D1%82%D0%B8%20%D0%BC%D0%B0%D1%81%D1%81%D0%BE%D0%B2%D0%BE%D0%B3%D0%BE%20%D0%BE%D0%B1%D1%81%D0%BB%D1%83%D0%B6%D0%B8%D0%B2%D0%B0%D0%BD%D0%B8%D1%8F%202021.pdf )

C бит/c – скорость обработки и передачи кадров
λ кадр/с – средняя скорость поступления кадров
1/µ бит/кадр – среднее экспоненциального распределение длины кадра

Рассмотрим среднее время задержки передачи кадра в нашей системе. Для начала рассмотрим его безотносительно введения каких-либо очередей передачи кадров (временных/частотных). Т.е. имеем дело с системой M/M/1. Тогда среднее время задержки будет равно произведению скорости обработки и передачи на среднюю длину кадра.

Скорость обработки и передачи с поправкой на длину одного кадра можно выразить как:

$$
\left| C - \frac{\lambda}{\mu} \right| \, \text{бит/с}
$$

Вычитаем из скорости обработки и передачи произведение средней скорости поступления и средней длины кадра, чтобы понять, насколько мы превышаем доступную скорость обработки и передачи (полагаем, что λ/µ бит/с в контексте нашей системы больше C бит/c, так как максимизируем скорость передачи)
Тогда задержку на один бит кадра можно описать как:

$$
t = \frac{1}{\left| C - \frac{\lambda}{\mu} \right|} \, \text{с/бит}
$$

Учитывая среднюю экспоненциального распределения длины кадров, получаем:

$$
T = t \, \text{с/бит} \cdot \frac{1}{\mu} \, \text{бит/кадр} = \frac{1}{\left| C - \frac{\lambda}{\mu} \right|} \cdot \frac{1}{\mu} = \frac{1}{\left| C \cdot \mu - \lambda \right|} \, \text{с/кадр}
$$

Теперь рассмотрим систему с некоторым разделением канала на «очереди». Т.е. работаем с системой M/M/N. Можем, к примеру, рассмотреть «очереди» в контексте частотного разделения канала на подканалы. Пусть мы делим доступный диапазон частот на N подканалов. Тогда согласно теореме Шеннона-Хартли скорость обработки и передачи уменьшится в N раз. $$
C = B \log_2\left(1 + \frac{S}{N}\right)
$$


где:

- \(C\) — пропускная способность канала, бит/с;
- \(B\) — полоса пропускания канала, Гц;
- \(S\) — полная мощность сигнала над полосой пропускания, Вт или В²;
- \(N\) — полная шумовая мощность над полосой пропускания, Вт или В²;
- \(S/N\) — отношение мощности сигнала к шуму (SNR).

B↓2 → C↓2.

При этом средняя скорость поступления кадров также уменьшится в N раз. На выходе получаем:
$$
T_{fdm} = \frac{1}{\left| \frac{C}{N} \cdot \mu - \frac{\lambda}{N} \right|} \, \text{с/кадр} = N \cdot T \, \text{с/кадр}
$$
